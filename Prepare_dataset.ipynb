```python
def prepare_dataset(source_path, target_path, train_size, validation_size, test_size):
    
    train_size, validation_size, test_size = train_size/100, validation_size/100, test_size/100
    
    images = [f for f in os.listdir(source_path) if os.path.isfile(os.path.join(source_path, f))]
    folder_names_part_1 = set()
    folder_names_part_2 = {"train_size", "validation_size", "test_size"}
    complete_folder_names = []

    for image in images:
        folder_names_part_1.add(image.split('.')[0])
    
    lenght=len(images)/len(folder_names_part_1)
    
    if ((lenght * train_size)%10)!=0 or  ((lenght*validation_size)%10)!=0 or ((lenght*test_size)%10)!=0:
        return("I can't split it like this")
    
    else: 

        for folder_name_p_1 in folder_names_part_1:
            for folder_name_p_2 in folder_names_part_2:
                new_path = os.path.join(target_path, "{}_{}".format(folder_name_p_1,folder_name_p_2))
                complete_folder_names.append("{}_{}".format(folder_name_p_1,folder_name_p_2))
                if not os.path.exists(new_path):
                    os.makedirs(new_path)

        cat_images = images [:int(len(images)/2)]
        len_cat = len(cat_images)

        dog_images = images [int(len(images)/2):]
        len_dog = len(dog_images)

        cat_train_size = cat_images[:int(len(cat_images)*train_size)]
        del cat_images[:len(cat_train_size)]
        cat_validation_size = cat_images[:int((len_cat*validation_size))]
        del cat_images[:len(cat_validation_size)]
        cat_test_size = cat_images

        dog_train_size = dog_images[:int(len(dog_images)*train_size)]
        del dog_images[:len(cat_train_size)]
        dog_validation_size = dog_images[:int((len_dog*validation_size))]
        del dog_images[:len(dog_validation_size)]
        dog_test_size = dog_images

        for name in cat_train_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[1])

        for name in cat_validation_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[0])

        for name in cat_test_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[2])

        for name in dog_train_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[4])

        for name in dog_validation_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[3])

        for name in dog_test_size:
            shutil.copy( source_path + "/" + name, target_path + "/" +  complete_folder_names[5])

        return "Done"
```
